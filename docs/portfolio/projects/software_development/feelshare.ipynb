# description_en

    ## Context 

    One day, I was discussing with my friend and he presents me his idea of android application to see the world feelings in real time. Vladimir is an artist (arts, music, photographer) and have no programming experience.
    He was very insterested in an application, that helps people to record their feelings and share them.
    The main goal of this app, is to quickly have an overview of the world feelings.

    ## Features expected

    * The user can log his feelings
    * View charts of user feeling
    * Map of the world with the world feelings 
    * Stats about the world feelings ( Ex : how many people are sad or happy today ?, ...)

    ## First release

    This app version is a rapid prototype of the idea and is not production ready.
    It helps me present the idea to other people and get their feedback about the project.

    * [Android - Play Store](https://play.google.com/store/apps/details?id=io.soglomania.appfeelshare)

    * [Github - Source code ](https://github.com/soglomania/feelshare/blob/master/README.md)
